```python
import numpy as np 
import matplotlib.pyplot as plt 
import sympy
import scipy.integrate as integrate
```

```python
def get_ar(n):
    res = []
    counter = 0
    for i in range(1, n+1):
        if sympy.isprime(i):
            counter += 1
        res += [counter]
    return res
```

```python
def plot_prime_counting_function(x, plot_functions=False, plot_log_int=False):
    ar = get_ar(x)

    if plot_functions:
        ar_0 = [a / np.log(a) if a != 1 else 0 for a in range(1, x + 1)]
        ar_1 = [a / (np.log(a)-1.08366) for a in range(1, x + 1)]
    if plot_log_int:
        ar_2_ = [integrate.quad(lambda x: 1/np.log(x), 2, a) for a in range(1, x + 1)]
        ar_2 = [x[0] for x in ar_2_]
    
    t1 = np.arange(1,x+1,1,dtype = int)
    
    plt.figure(figsize=(10,6))
    plt.step(t1, ar, label="Primzahlfunktion")
    if plot_functions:
        plt.plot(t1, ar_0, label=r"$\frac{x}{\ln(x)}$")
        plt.plot(t1, ar_1, label=r"$\frac{x}{\ln(x) - 1.08366}$")
        
    if plot_log_int:
        plt.plot(t1, ar_2, label=r"$\int_{2}^{x} \frac{\operatorname{d}t}{\ln(t)}$")
    plt.axis([0, x, 0, len(set(ar))]) 
    plt.legend(fontsize=14)
    plt.show()
```

```python
plot_prime_counting_function(100)
```

```python
plot_prime_counting_function(1000)
```
